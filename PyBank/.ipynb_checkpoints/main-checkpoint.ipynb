```python
# Initial Imports
import pandas as pd
from pathlib import Path
import csv
from csv import DictReader
import numpy as np

# Set filepath
filepath = Path("/Users/Brandon/Jupyter-Workspace/PyBank-Budget.csv")



# Open the file in "read" mode ('r') and store the contents in the variable 'file'
with open(filepath, 'r') as file:
    # Store all of the text from the file inside a variable called "text"
    # and print the contexts of the text file
    text = file.read()
    r = csv.reader(file)
    #print(text)

# @TODO: Parse the file line by line
    line_num = 1
    for line in file:
        print(f"line {line_num}: {line}")
        line_num += 1
        

# Sum profits and count months
with open(filepath) as file:
    headerline = file.readline()
    total = 0
    count = 0
    for row in csv.reader(file):
        #print(row)
        total += int(row[1])
        count += 1
    
    
# use DictReader to store Profit/Losses as a list
with open(filepath) as f:
    alist = [row['Profit/Losses'] for row in DictReader(f)]

# Store Profit/Losses as integers in list
profit_losses = []

for num in alist:
    profit_losses.append(int(num))

# Compute the profit change over period using numpy diff function
profit_change = np.diff(profit_losses)

# Convert array to list
profit_change_list = profit_change.tolist()

# Store max and min profit change
max_profit_change = max(profit_change_list)
min_profit_change = min(profit_change_list)

# Record index of max and min profit change to search for date
max_index = profit_change_list.index(max_profit_change)
min_index = profit_change_list.index(min_profit_change)
#print(max_index, min_index)

# Store average change
avg_change = sum(profit_change)/len(profit_change)


# Store date column as list:
with open(filepath) as f1:
    datelist = [row['Date'] for row in DictReader(f1)]

# Store the corresponding dates for max and min profit change
# adding 1 to index to allow for empty first row where profit change is null
max_profit_date = datelist[max_index + 1]
min_profit_date = datelist[min_index + 1]

print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {count}")
print(f"Total: ${total}")
print(f"Average Change: ${avg_change:.2f}")
print(f"Greatest Increase in Profits: {max_profit_date} (${max_profit_change})")
print(f"Greatest Decrease in Profits: {min_profit_date} (${min_profit_change})")



# Set output path
output_path = Path("/Users/Brandon/git/python-homework/PyBank/output.txt")

# Open the output path as a file object
with open(output_path, 'w') as file:
# Write financial analysis to file
    file.write("Financial Analysis\n")
    file.write("----------------------------\n")
    file.write(f"Total Months: {count}\n")
    file.write(f"Total: ${total}\n")
    file.write(f"Average Change: ${avg_change:.2f}\n")
    file.write(f"Greatest Increase in Profits: {max_profit_date} (${max_profit_change})\n")
    file.write(f"Greatest Decrease in Profits: {min_profit_date} (${min_profit_change})\n")
```
